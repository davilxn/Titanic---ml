Existem colunas com valores vazios na base de teste que nÃ£o estÃ£o vazios na base de treino. Dessa forma, faz-se necessÃ¡rio tratar a ausÃªncia de valores em ambas as bases.

### Tratamentos iniciais

Colunas com alta cardinalidade, isto Ã©, com elevado nÃºmero de categorias ou valores Ãºnicos (como o nome do passageiro ou o ticket, nesse caso), nÃ£o contribuem para a identificaÃ§Ã£o de padrÃµes significativos, uma vez que tornam o modelo mais especifista e menos generalista.
Inicialmente, colunas com alta cardinalidade serÃ£o descartadas. SerÃ£o consideradas, no entanto, para melhoria posterior do modelo.

OBS. A coluna PassengerId Ã© necessÃ¡ria ao submeter os resultados, por isso nÃ£o serÃ¡ excluÃ­da das bases de dados, mas nÃ£o serÃ¡ usada para treinar os modelos.

```python
# Analisando a cardinalidade das colunas
treino.nunique().sort_values(ascending=False)
```

```python
treino = treino.drop(["Name", "Ticket", "Cabin"], axis=1)
teste = teste.drop(["Name", "Ticket", "Cabin"], axis=1)
```

```python
# Substituindo valores nulos em "Age" pela mÃ©dia (a quantidade de valores nulos em amabas as bases Ã© inferior a 20% da quantidade de linhas)
media_idade_treino = treino.Age.mean()
media_idade_teste = teste.Age.mean()
treino.Age.fillna(media_idade_treino, inplace=True)
teste.Age.fillna(media_idade_teste, inplace=True)
```

```python
# Substuindo valores nulos em "Embarked" da base de treino pela moda
moda_embarque = treino.Embarked.mode()[0]
treino.Embarked.fillna(moda_embarque, inplace=True)
```

```python
# Substituindo valores nulos em "Fare" da base de teste pela mÃ©dia
media_tarifa = teste.Fare.mean()
teste.Fare.fillna(media_tarifa, inplace=True)
```

```python
treino.isnull().sum(), teste.isnull().sum()
```

## Utilizando, inicialmente, apenas colunas numÃ©ricas


```python
treino_num = treino.select_dtypes(exclude="object")
teste_num = teste.select_dtypes(exclude="object")
```

### SeleÃ§Ã£o de modelos para classificaÃ§Ã£o

- RegressÃ£o logÃ­stica (classificaÃ§Ã£o binÃ¡ria)
- SVM
- Random Forests

```python
from sklearn.model_selection import train_test_split

# Separando a base de treino em previsores (x) e classe (y)
x = treino_num.drop(["PassengerId", "Survived"], axis=1)
y = treino_num.Survived

# Separando em treino e validaÃ§Ã£o
x_treino, x_validacao, y_treino, y_validacao = train_test_split(x, y, test_size = 0.33, random_state=42) # Guia do mochileiro das galÃ¡xias
```

```python
# RegressÃ£o logÃ­stica
from sklearn.linear_model import LogisticRegression

modelo1 = LogisticRegression(random_state=42)
modelo1.fit(x_treino, y_treino)
pred1 = modelo1.predict(x_validacao)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

modelo2 = RandomForestClassifier(criterion="entropy", random_state=42)
modelo2.fit(x_treino, y_treino)
pred2 = modelo2.predict(x_validacao)
```

```python
# SVM
from sklearn.svm import SVC
modelo3 = SVC(C=2, kernel = "rbf", random_state=42)
modelo3.fit(x_treino, y_treino)
pred3 = modelo3.predict(x_validacao)
```

### AvaliaÃ§Ã£o dos modelos

```python
# Verificando a acurÃ¡cia dos modelos

from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix

# AcurÃ¡cia da RegressÃ£o LogÃ­stica
acc_mod1 = accuracy_score(y_validacao, pred1)

# AcurÃ¡cia da Random Forest
acc_mod2 = accuracy_score(y_validacao, pred2)

# AcurÃ¡cia do SVM
acc_mod3 = accuracy_score(y_validacao, pred3)

acc_mod1, acc_mod2, acc_mod3
```

```python
# Matriz de confusÃ£o da RegressÃ£o LogÃ­stica
cm = ConfusionMatrix(modelo1)
cm.fit(x_treino, y_treino)
cm.score(x_validacao, y_validacao)
```

```python
print(classification_report(y_validacao, pred1))
```

```python
# Matriz de confusÃ£o da Random Forest
cm = ConfusionMatrix(modelo2)
cm.fit(x_treino, y_treino)
cm.score(x_validacao, y_validacao)
```

```python
print(classification_report(y_validacao, pred2))
```

```python
# Matriz de confusÃ£o do SVM
cm = ConfusionMatrix(modelo3)
cm.fit(x_treino, y_treino)
cm.score(x_validacao, y_validacao)
```

```python
print(classification_report(y_validacao, pred3))
```

### Fazendo a previsÃ£o para os dados de teste

```python
# Obtendo a base de teste correta
x_teste = teste_num.drop(["PassengerId"], axis=1)
x_teste
```

```python
# SerÃ¡ utilizado para a previsÃ£o dos dados de teste o modelo que obteve melhor acurÃ¡cia: RegressÃ£o LogÃ­stica
pred_teste = modelo1.predict(x_teste)
```

```python
# Criando a nova coluna com a previsÃ£o na base de teste
teste["Survived"] = pred_teste
base_teste = teste[["PassengerId", "Survived"]]
base_teste
```
