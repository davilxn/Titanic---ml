Existem colunas com valores vazios na base de teste que nÃ£o estÃ£o vazios na base de treino. Dessa forma, faz-se necessÃ¡rio tratar a ausÃªncia de valores em ambas as bases.

## Tratamentos iniciais

Colunas com alta cardinalidade, isto Ã©, com elevado nÃºmero de categorias ou valores Ãºnicos (como o nome do passageiro ou o ticket, nesse caso), nÃ£o contribuem para a identificaÃ§Ã£o de padrÃµes significativos, uma vez que tornam o modelo mais especifista e menos generalista.
Inicialmente, colunas com alta cardinalidade serÃ£o descartadas. As mesmas serÃ£o consideradas, no entanto, para melhoria posterior do modelo.

OBS. A coluna PassengerId Ã© necessÃ¡ria ao submeter os resultados, por isso nÃ£o serÃ¡ excluÃ­da das bases de dados, mas nÃ£o serÃ¡ usada para treinar os modelos.

```python
# Analisando a cardinalidade das colunas
treino.nunique().sort_values(ascending=False)
```

```python
treino = treino.drop(["Name", "Ticket", "Cabin"], axis=1)
teste = teste.drop(["Name", "Ticket", "Cabin"], axis=1)
```

```python
# Substituindo valores nulos em "Age" pela mÃ©dia (a quantidade de valores nulos em amabas as bases Ã© inferior a 20% da quantidade de linhas)
media_idade_treino = treino.Age.mean()
media_idade_teste = teste.Age.mean()
treino.Age.fillna(media_idade_treino, inplace=True)
teste.Age.fillna(media_idade_teste, inplace=True)
```

```python
# Substuindo valores nulos em "Embarked" da base de treino pela moda
moda_embarque = treino.Embarked.mode()[0]
treino.Embarked.fillna(moda_embarque, inplace=True)
```

```python
# Substituindo valores nulos em "Fare" da base de teste pela mÃ©dia
media_tarifa = teste.Fare.mean()
teste.Fare.fillna(media_tarifa, inplace=True)
```

```python
treino.isnull().sum(), teste.isnull().sum()
```

### Tratando colunas categÃ³ricas (Sex e Embarked) com OneHotEncoder


```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Separando a base de treino em previsores (x) e classe (y)
x = treino.drop(["PassengerId", "Survived"], axis=1)
y = treino.Survived

# Removendo "PassengerId" da base de teste
x_teste = teste.drop(["PassengerId"], axis=1)

# Aplicando o OneHotEncoder na base de treino
categoricas = x.select_dtypes(include="object").columns
transformacao = [("OneHot", OneHotEncoder(), categoricas)]
onehot = ColumnTransformer(transformers=transformacao, remainder="passthrough")

x_onehot = onehot.fit_transform(x)
x_teste_onehot = onehot.fit_transform(x_teste)

# Somente para fins de visualizaÃ§Ã£o, serÃ£o reconstruÃ­dos os dataframes de treino e de teste com as novas colunas criadas pelo OneHotEncoder:

# ObtÃ©m os nomes das novas colunas criadas apÃ³s a codificaÃ§Ã£o
nomes_novas_colunas = onehot.named_transformers_['OneHot'].get_feature_names_out(categoricas)

# ObtÃ©m os nomes das colunas que nÃ£o foram codificadas
nomes_colunas_numericas = x.columns.difference(categoricas)

# Junta todas
novas_colunas_treino = list(nomes_novas_colunas) + list(nomes_colunas_numericas)

x = pd.DataFrame(x_onehot, columns=novas_colunas_treino)
x_teste = pd.DataFrame(x_teste_onehot, columns=novas_colunas)
```

### Escalonamento dos dados
DiferenÃ§as acentudas nas escalas dos dados entre as colunas (assim como a presenÃ§a de outliers) podem comprometer a estruturaÃ§Ã£o dos modelos, em especÃ­fico, os que realizam cÃ¡lculos de distÃ¢ncia, como o SVM.


```python
# Visualizando as escalas dos dados antes da padronizaÃ§Ã£o
x.boxplot(figsize=(13, 6))
plt.title('DistribuiÃ§Ã£o das Colunas')
plt.show()
```

As colunas "Fare" e "SibSp" tem escalas muito diferentes das escalas das outras colunas. AlÃ©m disso, a coluna "SibSp" possui outliers e, por esse fator, serÃ¡ realizada uma padronizaÃ§Ã£o dos dados com o RobustScaler.

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
x_scaler = scaler.fit_transform(x)
x_teste_scaler = scaler.fit_transform(x_teste)

# Voltando para o formato de dataframe
x = pd.DataFrame(data=x_scaler, columns=x.columns)
x_teste = pd.DataFrame(data=x_teste_scaler, columns=x_teste.columns)
```

```python
# Visualizando as escalas dos dados apÃ³s da padronizaÃ§Ã£o
x.boxplot(figsize=(13, 6))
plt.title('DistribuiÃ§Ã£o das Colunas')
plt.show()
```

## SeleÃ§Ã£o de atributos
Por testes anteriores, conclui-se que, a variÃ¢ncia mÃ©dia de todas as colunas Ã© relativamente alta, entÃ£o a seleÃ§Ã£o de atributos por Low Variance Ã© ineficaz.
SerÃ£o utilizados dois mÃ©todos em conjunto para estimar os melhores atributos da base de dados: teste de correlaÃ§Ã£o e avaliaÃ§Ã£o de importÃ¢ncia dos atributos por RandomForestClassifier.

```python
mat_corr = x.corr()
sns.heatmap(mat_corr, annot=True)
```

```python
threshold = 0.8

# Encontrar pares de atributos altamente correlacionados
pares_corr_alta = [(i, j) for i in range(mat_corr.shape[0]) for j in range(i+1, mat_corr.shape[1]) if abs(mat_corr.iloc[i, j]) > threshold]

# Exibir os pares de atributos altamente correlacionados
print("Pares de atributos altamente correlacionados:", pares_corr_alta)
```

```python
# SeleÃ§Ã£o por RandomForestClassifier

from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(x, y)

# Obter a importÃ¢ncia dos atributos
importancias = rf_model.feature_importances_

df_import = pd.DataFrame({'Atributo': x.columns, 'ImportÃ¢ncia': importancias})

# Ordenar por importÃ¢ncia
df_import = df_import.sort_values(by='ImportÃ¢ncia', ascending=False)

# Exibir a importÃ¢ncia dos atributos
print(df_import)

# Criar um grÃ¡fico de barras para visualizar a importÃ¢ncia dos atributos
plt.figure(figsize=(10, 6))
plt.bar(df_import['Atributo'], df_import['ImportÃ¢ncia'])
plt.title('ImportÃ¢ncia dos Atributos - Random Forest')
plt.xlabel('Atributo')
plt.ylabel('ImportÃ¢ncia')
plt.show()
```

Para fins de teste, desta maneira, serÃ£o excluÃ­dos os atributos "Sex_male" em decorrÃªncia da correlaÃ§Ã£o igual a -1 com "Sex_female" e o atributo "Embarked_Q" e observaremos os impactos na classificaÃ§Ã£o.

```python
x = x.drop(["Sex_male"], axis=1)
x_teste = x_teste.drop(["Sex_male"], axis=1)
```

```python
# RegressÃ£o logÃ­stica
modelo1 = LogisticRegression(C=1, solver="saga", tol=0.0001, random_state=42, max_iter=250)
modelo1.fit(x_treino, y_treino)
pred1 = modelo1.predict(x_validacao)
```

```python
# Multilayer Perceptron (MLP)
modelo2 = MLPClassifier(max_iter = 2000, verbose=False, tol=0.00001, activation="relu", batch_size=10, solver="sgd")
modelo2.fit(x_treino, y_treino)
pred2 = modelo2.predict(x_validacao)
```

```python
# SVM
modelo3 = SVC(C=100, kernel = "poly", tol= 0.001, random_state=42)
modelo3.fit(x_treino, y_treino)
pred3 = modelo3.predict(x_validacao)
```
