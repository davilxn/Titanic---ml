- SibSp: Número de irmãos/cônjuges a bordo
- Parch: Número de pais/filhos a bordo

Juntas, a soma dos valores dessas colunas representam a quantidade total de acompanhantes (familiares) de um determinado passageiro.

É possível concluir que pessoas acompanhadas com até 3 familiares tiveram maior chance de sobrevivência.

# Modelo extra - Torch ANN

### Ajustando e dividindo dados

```python
import torch
# Transformando dados em tensores
X = torch.from_numpy(X.to_numpy()).type(torch.float32)
y = torch.from_numpy(y.to_numpy()).type(torch.float32)
X_teste = torch.from_numpy(X_teste.to_numpy()).type(torch.float32)

# Seeds
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
torch.cuda.manual_seed(RANDOM_SEED)

# Ajuste de device
device = "cuda" if torch.cuda.is_available() else "cpu"

# Train-test-split
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=RANDOM_SEED)
```

### Construindo o modelo

```python
from torch import nn

# Construindo o modelo
class Titanic(nn.Module):
  def __init__(self, input_feat, output_feat, hidden_units):
    super().__init__()
    self.input_layer = nn.Linear(in_features=input_feat, out_features=hidden_units)
    self.hidden_layers = nn.Sequential(
        nn.Linear(in_features=hidden_units, out_features=hidden_units),
        nn.ReLU(),
        nn.Linear(in_features=hidden_units, out_features=hidden_units),
        nn.ReLU(),
        nn.Linear(in_features=hidden_units, out_features=hidden_units),
        nn.Sigmoid(),
        nn.Linear(in_features=hidden_units, out_features=hidden_units),
        nn.Sigmoid(),
        nn.Linear(in_features=hidden_units, out_features=hidden_units),
        nn.ReLU(),
        nn.Linear(in_features=hidden_units, out_features=hidden_units),
        nn.ReLU(),
        nn.Linear(in_features=hidden_units, out_features=hidden_units),
        nn.ReLU(),
    )
    self.output_layer = nn.Linear(in_features=hidden_units, out_features=output_feat)

  def forward(self, X:torch.Tensor) -> torch.Tensor:
    x = torch.relu(self.input_layer(X))
    x = self.hidden_layers(x)
    x = self.output_layer(x)
    return x

# Instanciando o modelo
model_0 = Titanic(input_feat=X_train.shape[1], output_feat=1, hidden_units=32).to(device)
next(model_0.parameters()).device
```

```python
# Função de perda e otimizador

loss_fn = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(params=model_0.parameters(), lr=0.003)
```

```python
# Função de acurácia de treino e validação
def acc_fn(y_true, y_pred):
  correct = torch.eq(y_true, y_pred).sum().item()
  acc = (correct / len(y_pred)) * 100
  return acc
```

### Treinamento e teste

```python
# Loop de treino e de teste

# Seeds
torch.manual_seed(RANDOM_SEED)
torch.cuda.manual_seed(RANDOM_SEED)

# Épocas
epochs = 1000

# Dados
X_train, y_train = X_train.to(device), y_train.to(device)
X_val, y_val = X_val.to(device), y_val.to(device)

# Loop
for epoch in range(epochs):
  ### Treinamento
  model_0.train()

  # 1. Forward
  y_logits = model_0(X_train).squeeze()
  y_preds = torch.round(torch.sigmoid(y_logits))

  # 2. Cálculo da perda
  train_loss = loss_fn(y_logits, y_train)
  train_acc = acc_fn(y_train, y_preds)

  # 3. Zerando gradientes
  optimizer.zero_grad()

  # 4. Backpropagation
  train_loss.backward()

  # Atualização de pesos
  optimizer.step()

  ### Teste/Validação
  model_0.eval()
  with torch.inference_mode():
    test_logits = model_0(X_val).squeeze()
    test_preds = torch.round(torch.sigmoid(test_logits))

    val_loss = loss_fn(test_logits, y_val)
    val_acc = acc_fn(y_val, test_preds)

    if epoch % 100 == 0:
      print(f"Epoch: {epoch} | Train Loss: {train_loss:.5f} | Train Acc: {train_acc:.2f}% | Val Loss: {val_loss:.5f} | Val Acc: {val_acc:.2f}%")

    if epoch >= 500 and val_acc > 82:
      print(f"Acc: {val_acc}. Early stopping")
      break
```
