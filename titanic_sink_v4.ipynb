## Tratamentos iniciais

Colunas com alta cardinalidade, isto é, com elevado número de categorias ou valores únicos (como o nome do passageiro ou o ticket, nesse caso), não contribuem para a identificação de padrões significativos, uma vez que tornam o modelo mais especifista e menos generalista.
Inicialmente, colunas com alta cardinalidade serão descartadas. As mesmas serão consideradas, no entanto, para melhoria posterior do modelo.

OBS. A coluna PassengerId é necessária ao submeter os resultados, por isso não será excluída das bases de dados, mas não será usada para treinar os modelos.

```python
# Analisando a cardinalidade das colunas
treino.nunique().sort_values(ascending=False)
```

```python
treino = treino.drop(["Name", "Ticket", "Cabin"], axis=1)
teste = teste.drop(["Name", "Ticket", "Cabin"], axis=1)
```

```python
# Substituindo valores nulos em "Age" pela média (a quantidade de valores nulos em amabas as bases é inferior a 20% da quantidade de linhas)
media_idade_treino = treino.Age.mean()
media_idade_teste = teste.Age.mean()
treino.Age.fillna(media_idade_treino, inplace=True)
teste.Age.fillna(media_idade_teste, inplace=True)
```

```python
# Substuindo valores nulos em "Embarked" da base de treino pela moda
moda_embarque = treino.Embarked.mode()[0]
treino.Embarked.fillna(moda_embarque, inplace=True)
```

```python
# Substituindo valores nulos em "Fare" da base de teste pela média
media_tarifa = teste.Fare.mean()
teste.Fare.fillna(media_tarifa, inplace=True)
```

```python
treino.isnull().sum(), teste.isnull().sum()
```

### Tratando colunas categóricas (Sex e Embarked) com OneHotEncoder


```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Separando a base de treino em previsores (x) e classe (y)
x = treino.drop(["PassengerId", "Survived"], axis=1)
y = treino.Survived

# Removendo "PassengerId" da base de teste
x_teste = teste.drop(["PassengerId"], axis=1)

# Aplicando o OneHotEncoder na base de treino
categoricas = x.select_dtypes(include="object").columns
transformacao = [("OneHot", OneHotEncoder(), categoricas)]
onehot = ColumnTransformer(transformers=transformacao, remainder="passthrough")

x_onehot = onehot.fit_transform(x)
x_teste_onehot = onehot.fit_transform(x_teste)

# Somente para fins de visualização, serão reconstruídos os dataframes de treino e de teste com as novas colunas criadas pelo OneHotEncoder:

# Obtém os nomes das novas colunas criadas após a codificação
nomes_novas_colunas = onehot.named_transformers_['OneHot'].get_feature_names_out(categoricas)

# Obtém os nomes das colunas que não foram codificadas
nomes_colunas_numericas = x.columns.difference(categoricas)

# Junta todas
novas_colunas_treino = list(nomes_novas_colunas) + list(nomes_colunas_numericas)

x = pd.DataFrame(x_onehot, columns=novas_colunas_treino)
x_teste = pd.DataFrame(x_teste_onehot, columns=novas_colunas)
```

### Escalonamento dos dados
Diferenças acentudas nas escalas dos dados entre as colunas (assim como a presença de outliers) podem comprometer a estruturação dos modelos, em específico, os que realizam cálculos de distância, como o SVM.


```python
# Visualizando as escalas dos dados antes da padronização
x.boxplot(figsize=(13, 6))
plt.title('Distribuição das Colunas')
plt.show()
```

As colunas "Fare" e "SibSp" tem escalas muito diferentes das escalas das outras colunas. Além disso, a coluna "SibSp" possui outliers e, por esse fator, será realizada uma padronização dos dados com o RobustScaler.

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
x_scaler = scaler.fit_transform(x)
x_teste_scaler = scaler.fit_transform(x_teste)

# Voltando para o formato de dataframe
x = pd.DataFrame(data=x_scaler, columns=x.columns)
x_teste = pd.DataFrame(data=x_teste_scaler, columns=x_teste.columns)
```

```python
# Visualizando as escalas dos dados após da padronização
x.boxplot(figsize=(13, 6))
plt.title('Distribuição das Colunas')
plt.show()
```

## Seleção de atributos
Por testes anteriores, conclui-se que, a variância média de todas as colunas é relativamente alta, então a seleção de atributos por Low Variance é ineficaz.
Serão utilizados dois métodos em conjunto para estimar os melhores atributos da base de dados: teste de correlação e avaliação de importância dos atributos por RandomForestClassifier.

```python
mat_corr = x.corr()
sns.heatmap(mat_corr, annot=True)
```

```python
threshold = 0.8

# Encontrar pares de atributos altamente correlacionados
pares_corr_alta = [(i, j) for i in range(mat_corr.shape[0]) for j in range(i+1, mat_corr.shape[1]) if abs(mat_corr.iloc[i, j]) > threshold]

# Exibir os pares de atributos altamente correlacionados
print("Pares de atributos altamente correlacionados:", pares_corr_alta)
```

```python
# Seleção por RandomForestClassifier

from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(x, y)

# Obter a importância dos atributos
importancias = rf_model.feature_importances_

df_import = pd.DataFrame({'Atributo': x.columns, 'Importância': importancias})

# Ordenar por importância
df_import = df_import.sort_values(by='Importância', ascending=False)

# Exibir a importância dos atributos
print(df_import)

# Criar um gráfico de barras para visualizar a importância dos atributos
plt.figure(figsize=(10, 6))
plt.bar(df_import['Atributo'], df_import['Importância'])
plt.title('Importância dos Atributos - Random Forest')
plt.xlabel('Atributo')
plt.ylabel('Importância')
plt.show()
```

Para fins de teste, desta maneira, serão excluídos os atributos "Sex_male" em decorrência da correlação igual a -1 com "Sex_female" e o atributo "Embarked_Q" e observaremos os impactos na classificação.

```python
x = x.drop(["Sex_male"], axis=1)
x_teste = x_teste.drop(["Sex_male"], axis=1)
```

## Seleção de modelos para classificação

- Regressão logística (classificação binária)
- SVM

```python
# Aplicação do GridSearch - RandomForest
parametros = {"criterion": ['gini', 'entropy'],
              "n_estimators": [10,40,100,150],
              "min_samples_split": [2,5,10],
              "min_samples_leaf":[1,5,10]
              }

grid_search = GridSearchCV(estimator=RandomForestClassifier(), param_grid=parametros)
grid_search.fit(x_treino, y_treino)
melhores_parametros = grid_search.best_params_

melhores_parametros
```

```python
# Regressão logística
modelo1 = LogisticRegression(C=1, solver="saga", tol=0.0001, random_state=42, max_iter=250)
modelo1.fit(x_treino, y_treino)
pred1 = modelo1.predict(x_validacao)
```

```python
# Multilayer Perceptron (MLP)
modelo2 = MLPClassifier(max_iter = 2000, verbose=False, tol=0.00001, activation="relu", batch_size=10, solver="sgd")
modelo2.fit(x_treino, y_treino)
pred2 = modelo2.predict(x_validacao)
```

```python
# SVM
modelo3 = SVC(C=100, kernel = "poly", tol= 0.001, random_state=42)
modelo3.fit(x_treino, y_treino)
pred3 = modelo3.predict(x_validacao)
```

```python
# RandomForest
modelo4 = RandomForestClassifier(criterion="entropy", min_samples_leaf=5, min_samples_split=2, n_estimators=40)
modelo4.fit(x_treino, y_treino)
pred4 = modelo4.predict(x_validacao)
```

### Avaliação dos modelos

```python
# Verificando a acurácia dos modelos

from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix

# Acurácia da Regressão Logística
acc_mod1 = accuracy_score(y_validacao, pred1)

# Acurácia da MLP
acc_mod2 = accuracy_score(y_validacao, pred2)

# Acurácia do SVM
acc_mod3 = accuracy_score(y_validacao, pred3)

# Acurácia do RandomForest
acc_mod4 = accuracy_score(y_validacao, pred4)

acc_mod1, acc_mod2, acc_mod3, acc_mod4
```
