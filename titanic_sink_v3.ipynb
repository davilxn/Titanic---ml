# Tratamento dos dados

### Tratamentos iniciais

Colunas com alta cardinalidade, isto é, com elevado número de categorias ou valores únicos (como o nome do passageiro ou o ticket, nesse caso), não contribuem para a identificação de padrões significativos, uma vez que tornam o modelo mais especifista e menos generalista.
Inicialmente, colunas com alta cardinalidade serão descartadas. As mesmas serão consideradas, no entanto, para melhoria posterior do modelo.
'
OBS. A coluna PassengerId é necessária ao submeter os resultados, por isso não será excluída das bases de dados, mas não será usada para treinar os modelos.

```python
# Analisando a cardinalidade das colunas
treino.nunique().sort_values(ascending=False)
```

```python
# Substituindo valores nulos em "Age" pela média (a quantidade de valores nulos em amabas as bases é inferior a 20% da quantidade de linhas)
media_idade_treino = treino.Age.mean()
media_idade_teste = teste.Age.mean()
treino.Age.fillna(media_idade_treino, inplace=True)
teste.Age.fillna(media_idade_teste, inplace=True)
```

```python
# Substuindo valores nulos em "Embarked" da base de treino pela moda
moda_embarque = treino.Embarked.mode()[0]
treino.Embarked.fillna(moda_embarque, inplace=True)
```

```python
# Substituindo valores nulos em "Fare" da base de teste pela média
media_tarifa = teste.Fare.mean()
teste.Fare.fillna(media_tarifa, inplace=True)
```

# Análise de atributos



### As colunas "SibSp" e "Parch"
- SibSp: Número de irmãos/cônjuges a bordo
- Parch: Número de pais/filhos a bordo

Juntas, a soma dos valores dessas colunas representam a quantidade total de acompanhantes (familiares) de um determinado passageiro.

```python
# Treino
family_treino = treino.SibSp + treino.Parch
treino["Family"] = family_treino

# Teste
family_teste = teste.SibSp + teste.Parch
teste["Family"] = family_teste

family_data = {"Sobrev": treino.groupby("Family")["Survived"].sum(),
          "Qtd_total": treino.groupby("Family")["Survived"].count(),
          "Taxa_sobrev": treino.groupby("Family")["Survived"].mean()}

family_df = pd.DataFrame(data=family_data)
family_df
```

É possível concluir que pessoas acompanhadas com até 3 familiares tiveram maior chance de sobrevivência.

```python
# Tratamento da coluna "Family"

# Função condicional
def membros(x):
    if x == 0 or x > 3:
        return 0
    else:
        return 1

# Aplicando a função condicional
treino["Family"] = treino["Family"].apply(membros)
teste["Family"] = teste["Family"].apply(membros)
```

### A coluna "Sex"

```python
sex_data = {"Sobrev": treino.groupby("Sex")["Survived"].sum(),
            "Qtd_total": treino.groupby("Sex")["Survived"].count(),
            "Taxa_sobrev": treino.groupby("Sex")["Survived"].mean()}

df_sex = pd.DataFrame(data=sex_data)
df_sex
```

Mulheres tiveram maior taxa de sobrevivência. A ordem de entrada nos botes salva-vidas durante o abandono do navio era de que mulheres e crianças adentrassem primeiro.

```python
# Tratamento da coluna "Sex"

# Função condicional
def sexo(x):
    if x == "male":
        return 0
    else:
        return 1

# Aplicando a função condicional
treino["Sex"] = treino["Sex"].apply(sexo)
teste["Sex"] = teste["Sex"].apply(sexo)
```

A idade dos passageiros influencia diretamente em suas chances de sobrevivência.

Convém agrupar as idades em intervalos e analisar a taxa de sobrevivência.

```python
ages_inter = pd.cut(treino['Age'], 6)
ages_inter.value_counts()
```

```python
# Categorizando a coluna "Age", de baixa cardinalidade para uma coluna numérica
def ages(x):
    if x >= 0 and x < 13:
        return 0
    elif x >= 13 and x < 26:
        return 1
    elif x >= 26 and x < 40:
        return 2
    elif x >= 40 and x < 53:
        return 3
    elif x >= 53 and x < 66:
        return 4
    elif x >= 66 and x <= 80:
        return 5

treino["Ages_cat"] = treino.Age.apply(ages)
teste["Ages_cat"] = teste.Age.apply(ages)
```

```python
age_data = {"Sobrev": treino.groupby("Ages_cat")["Survived"].sum(),
            "Qtd_total": treino.groupby("Ages_cat")["Survived"].count(),
            "Taxa_sobrev": treino.groupby("Ages_cat")["Survived"].mean()}

df_age = pd.DataFrame(data=age_data)
df_age
```

Na presença de assimetria positiva e outliers, é válido testar uma transformação logarítimica na coluna e verificar a correlação com a taxa de sobrevivência, além de tentar agrupar os os valores da coluna em intervalos (discretização).

Transformação logaritmica

```python
log_fare_treino = np.log(treino.Fare + 1)
log_fare_teste = np.log(teste.Fare + 1)

sns.histplot(log_fare_treino, kde=True)
```

```python
log_fare_treino.corr(treino.Survived)
```

Binning (discretização)

# Enconding de colunas categóricas e escalonamento dos dados

### Tratando a coluna Embarked com OneHotEncoder


```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Separando a base de treino em previsores (x) e classe (y)
x = treino.drop(["PassengerId", "Survived"], axis=1)
y = treino.Survived

# Removendo "PassengerId" da base de teste
x_teste = teste.drop(["PassengerId"], axis=1)

# Aplicando o OneHotEncoder na base de treino
transformacao = [("OneHot", OneHotEncoder(), ["Embarked"])]
onehot = ColumnTransformer(transformers=transformacao, remainder="passthrough")

# Ajustando e transformando a base de treino
x_onehot = onehot.fit_transform(x)
x_teste_onehot = onehot.transform(x_teste)

# Recuperando os nomes das novas colunas
onehot_feature_names = onehot.named_transformers_["OneHot"].get_feature_names_out(["Embarked"])

# Juntando os nomes das novas colunas com as colunas que permaneceram inalteradas
new_columns = list(onehot_feature_names) + [col for col in x.columns if col != "Embarked"]

# Convertendo o array transformado de volta para DataFrame
x_onehot_df = pd.DataFrame(x_onehot, columns=new_columns)
x_teste_onehot_df = pd.DataFrame(x_teste_onehot, columns=new_columns)
x_teste_onehot_df
```

# Seleção de atributos

Por testes anteriores, conclui-se que a variância média de todas as colunas é relativamente alta, então a seleção de atributos por Low Variance é ineficaz.
Serão utilizados dois métodos em conjunto para estimar os melhores atributos da base de dados: teste de correlação e avaliação de importância dos atributos por RandomForestClassifier.

```python
X = x_onehot_df
X_teste = x_teste_onehot_df
sns.heatmap(X.corr(), annot=True)
```

```python
threshold = 0.75
corr = X.corr()
# Encontrar pares de atributos altamente correlacionados
pares_corr_alta = [(i, j) for i in range(corr.shape[0]) for j in range(i+1, corr.shape[1]) if abs(corr.iloc[i, j]) >= threshold]

# Exibir os pares de atributos altamente correlacionados
print("Pares de atributos altamente correlacionados:", pares_corr_alta)
```

```python
# Seleção por RandomForestClassifier

from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X, y)

# Obter a importância dos atributos
importancias = rf_model.feature_importances_

df_import = pd.DataFrame({'Atributo': X.columns, 'Importância': importancias})

# Ordenar por importância
df_import = df_import.sort_values(by='Importância', ascending=False)

# Exibir a importância dos atributos
print(df_import)

# Criar um gráfico de barras para visualizar a importância dos atributos
plt.figure(figsize=(10, 6))
plt.bar(df_import['Atributo'], df_import['Importância'])
plt.title('Importância dos Atributos - Random Forest')
plt.xlabel('Atributo')
plt.ylabel('Importância')
plt.show()
```

A coluna recém criada "Log Fare" apresenta maior correlação com a variável alvo "Survived" e foi dada como a segunda mais importante para a seleção de atributos por Random Forest.
Desta forma, a coluna "Fare Interv" será excluída.

```python
X.drop(["Fare_Interv", "Fare"], axis=1, inplace=True)
X_teste.drop(["Fare_Interv", "Fare"], axis=1, inplace=True)
```

# Seleção e treinamento de modelos para classificação

- Regressão logística
- SVM
- Random Forests

```python
# Separando em treino e validação
from sklearn.model_selection import train_test_split

X_treino, X_validacao, y_treino, y_validacao = train_test_split(X, y, test_size = 0.33, random_state=42)
```

```python
# Regressão logística
from sklearn.linear_model import LogisticRegression

modelo1 = LogisticRegression(random_state=42, max_iter=250)
modelo1.fit(X_treino, y_treino)
pred1 = modelo1.predict(X_validacao)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
modelo2 = RandomForestClassifier(criterion="entropy", random_state=42)
modelo2.fit(X_treino, y_treino)
pred2 = modelo2.predict(X_validacao)
```

```python
# SVM
from sklearn.svm import SVC
modelo3 = SVC(C=15, kernel='rbf', tol= 0.0001, max_iter=2000, random_state=42)
modelo3.fit(X_treino, y_treino)
pred3 = modelo3.predict(X_validacao)
```

# Avaliação dos modelos

```python
# Verificando a acurácia dos modelos

from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix

# Acurácia da Regressão Logística
acc_mod1 = accuracy_score(y_validacao, pred1)

# Acurácia da Random Forest
acc_mod2 = accuracy_score(y_validacao, pred2)

# Acurácia do SVM
acc_mod3 = accuracy_score(y_validacao, pred3)

acc_mod1, acc_mod2, acc_mod3
```

```python
print(classification_report(y_validacao, pred2))
```

```python
print(classification_report(y_validacao, pred3))
```
