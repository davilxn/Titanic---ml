# Tratamento dos dados

### Tratamentos iniciais

Colunas com alta cardinalidade, isto Ã©, com elevado nÃºmero de categorias ou valores Ãºnicos (como o nome do passageiro ou o ticket, nesse caso), nÃ£o contribuem para a identificaÃ§Ã£o de padrÃµes significativos, uma vez que tornam o modelo mais especifista e menos generalista.
Inicialmente, colunas com alta cardinalidade serÃ£o descartadas. As mesmas serÃ£o consideradas, no entanto, para melhoria posterior do modelo.
'
OBS. A coluna PassengerId Ã© necessÃ¡ria ao submeter os resultados, por isso nÃ£o serÃ¡ excluÃ­da das bases de dados, mas nÃ£o serÃ¡ usada para treinar os modelos.

```python
# Analisando a cardinalidade das colunas
treino.nunique().sort_values(ascending=False)
```

```python
# Substituindo valores nulos em "Age" pela mÃ©dia (a quantidade de valores nulos em amabas as bases Ã© inferior a 20% da quantidade de linhas)
media_idade_treino = treino.Age.mean()
media_idade_teste = teste.Age.mean()
treino.Age.fillna(media_idade_treino, inplace=True)
teste.Age.fillna(media_idade_teste, inplace=True)
```

```python
# Substuindo valores nulos em "Embarked" da base de treino pela moda
moda_embarque = treino.Embarked.mode()[0]
treino.Embarked.fillna(moda_embarque, inplace=True)
```

```python
# Substituindo valores nulos em "Fare" da base de teste pela mÃ©dia
media_tarifa = teste.Fare.mean()
teste.Fare.fillna(media_tarifa, inplace=True)
```

# AnÃ¡lise de atributos



### As colunas "SibSp" e "Parch"
- SibSp: NÃºmero de irmÃ£os/cÃ´njuges a bordo
- Parch: NÃºmero de pais/filhos a bordo

Juntas, a soma dos valores dessas colunas representam a quantidade total de acompanhantes (familiares) de um determinado passageiro.

```python
# Treino
family_treino = treino.SibSp + treino.Parch
treino["Family"] = family_treino

# Teste
family_teste = teste.SibSp + teste.Parch
teste["Family"] = family_teste

family_data = {"Sobrev": treino.groupby("Family")["Survived"].sum(),
          "Qtd_total": treino.groupby("Family")["Survived"].count(),
          "Taxa_sobrev": treino.groupby("Family")["Survived"].mean()}

family_df = pd.DataFrame(data=family_data)
family_df
```

Ã‰ possÃ­vel concluir que pessoas acompanhadas com atÃ© 3 familiares tiveram maior chance de sobrevivÃªncia.

```python
# Tratamento da coluna "Family"

# FunÃ§Ã£o condicional
def membros(x):
    if x == 0 or x > 3:
        return 0
    else:
        return 1

# Aplicando a funÃ§Ã£o condicional
treino["Family"] = treino["Family"].apply(membros)
teste["Family"] = teste["Family"].apply(membros)
```

### A coluna "Sex"

```python
sex_data = {"Sobrev": treino.groupby("Sex")["Survived"].sum(),
            "Qtd_total": treino.groupby("Sex")["Survived"].count(),
            "Taxa_sobrev": treino.groupby("Sex")["Survived"].mean()}

df_sex = pd.DataFrame(data=sex_data)
df_sex
```

Mulheres tiveram maior taxa de sobrevivÃªncia. A ordem de entrada nos botes salva-vidas durante o abandono do navio era de que mulheres e crianÃ§as adentrassem primeiro.

```python
# Tratamento da coluna "Sex"

# FunÃ§Ã£o condicional
def sexo(x):
    if x == "male":
        return 0
    else:
        return 1

# Aplicando a funÃ§Ã£o condicional
treino["Sex"] = treino["Sex"].apply(sexo)
teste["Sex"] = teste["Sex"].apply(sexo)
```

ConvÃ©m agrupar as idades em intervalos e analisar a taxa de sobrevivÃªncia.

```python
ages_inter = pd.cut(treino['Age'], 6)
ages_inter.value_counts()
```

```python
# Categorizando a coluna "Age", de baixa cardinalidade para uma coluna numÃ©rica
def ages(x):
    if x >= 0 and x < 13:
        return 0
    elif x >= 13 and x < 26:
        return 1
    elif x >= 26 and x < 40:
        return 2
    elif x >= 40 and x < 53:
        return 3
    elif x >= 53 and x < 66:
        return 4
    elif x >= 66 and x <= 80:
        return 5

treino["Ages_cat"] = treino.Age.apply(ages)
teste["Ages_cat"] = teste.Age.apply(ages)
```

```python
age_data = {"Sobrev": treino.groupby("Ages_cat")["Survived"].sum(),
            "Qtd_total": treino.groupby("Ages_cat")["Survived"].count(),
            "Taxa_sobrev": treino.groupby("Ages_cat")["Survived"].mean()}

df_age = pd.DataFrame(data=age_data)
df_age
```

Na presenÃ§a de assimetria positiva e outliers, Ã© vÃ¡lido testar uma transformaÃ§Ã£o logarÃ­timica na coluna e verificar a correlaÃ§Ã£o com a taxa de sobrevivÃªncia, alÃ©m de tentar agrupar os os valores da coluna em intervalos (discretizaÃ§Ã£o).

TransformaÃ§Ã£o logaritmica

```python
log_fare_treino = np.log(treino.Fare + 1)
log_fare_teste = np.log(teste.Fare + 1)

sns.histplot(log_fare_treino, kde=True)
```

```python
log_fare_treino.corr(treino.Survived)
```

Binning (discretizaÃ§Ã£o)

```python
fares_inter = pd.cut(treino['Fare'], 6)
fares_inter.value_counts()
```

```python
def fares(x):
  if x >= 0 and x < 85:
    return 0
  elif x >= 85 and x < 170:
    return 1
  elif x >= 170 and x < 256:
    return 2
  else:
    return 3

fare_inter_treino = treino["Fare"].apply(fares)
fare_inter_teste = teste["Fare"].apply(fares)
```

```python
# Verificando correlaÃ§Ãµes com a variÃ¡vel alvo
log_fare_treino.corr(treino.Survived), fare_inter_treino.corr(treino.Survived), treino.Fare.corr(treino.Survived)
```

AlÃ©m de verificaÃ§Ãµes empÃ­ricas de acurÃ¡cia, serÃ¡ realizada uma seleÃ§Ã£o de atributos por mÃ©todos baseados em modelos para decidir qual ou quais dessas variÃ¡veis sÃ£o mais Ãºteis para o treinamento.

```python
treino["Log_Fare"] = log_fare_treino
teste["Log_Fare"] = log_fare_teste
treino["Fare_Interv"] = fare_inter_treino
teste["Fare_Interv"] = fare_inter_teste
```

```python
def age_pclass(x):
  if x > 0 and x < 3:
    return 0
  elif x >= 3 and x < 6:
    return 1
  elif x >= 6 and x < 9:
    return 2
  elif x >= 9 and x < 12:
    return 3
  else:
    return 4

treino["AgePclass"] = treino["AgePclass"].apply(age_pclass)
```

# Enconding de colunas categÃ³ricas e escalonamento dos dados

### Tratando a coluna Embarked com OneHotEncoder


```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Separando a base de treino em previsores (x) e classe (y)
x = treino.drop(["PassengerId", "Survived"], axis=1)
y = treino.Survived

# Removendo "PassengerId" da base de teste
x_teste = teste.drop(["PassengerId"], axis=1)

# Aplicando o OneHotEncoder na base de treino
transformacao = [("OneHot", OneHotEncoder(), ["Embarked"])]
onehot = ColumnTransformer(transformers=transformacao, remainder="passthrough")

# Ajustando e transformando a base de treino
x_onehot = onehot.fit_transform(x)
x_teste_onehot = onehot.transform(x_teste)

# Recuperando os nomes das novas colunas
onehot_feature_names = onehot.named_transformers_["OneHot"].get_feature_names_out(["Embarked"])

# Juntando os nomes das novas colunas com as colunas que permaneceram inalteradas
new_columns = list(onehot_feature_names) + [col for col in x.columns if col != "Embarked"]

# Convertendo o array transformado de volta para DataFrame
x_onehot_df = pd.DataFrame(x_onehot, columns=new_columns)
x_teste_onehot_df = pd.DataFrame(x_teste_onehot, columns=new_columns)
x_teste_onehot_df
```

# SeleÃ§Ã£o de atributos

Por testes anteriores, conclui-se que a variÃ¢ncia mÃ©dia de todas as colunas Ã© relativamente alta, entÃ£o a seleÃ§Ã£o de atributos por Low Variance Ã© ineficaz.
SerÃ£o utilizados dois mÃ©todos em conjunto para estimar os melhores atributos da base de dados: teste de correlaÃ§Ã£o e avaliaÃ§Ã£o de importÃ¢ncia dos atributos por RandomForestClassifier.

```python
X = x_onehot_df
X_teste = x_teste_onehot_df
sns.heatmap(X.corr(), annot=True)
```

```python
threshold = 0.75
corr = X.corr()
# Encontrar pares de atributos altamente correlacionados
pares_corr_alta = [(i, j) for i in range(corr.shape[0]) for j in range(i+1, corr.shape[1]) if abs(corr.iloc[i, j]) >= threshold]

# Exibir os pares de atributos altamente correlacionados
print("Pares de atributos altamente correlacionados:", pares_corr_alta)
```

```python
# SeleÃ§Ã£o por RandomForestClassifier

from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X, y)

# Obter a importÃ¢ncia dos atributos
importancias = rf_model.feature_importances_

df_import = pd.DataFrame({'Atributo': X.columns, 'ImportÃ¢ncia': importancias})

# Ordenar por importÃ¢ncia
df_import = df_import.sort_values(by='ImportÃ¢ncia', ascending=False)

# Exibir a importÃ¢ncia dos atributos
print(df_import)

# Criar um grÃ¡fico de barras para visualizar a importÃ¢ncia dos atributos
plt.figure(figsize=(10, 6))
plt.bar(df_import['Atributo'], df_import['ImportÃ¢ncia'])
plt.title('ImportÃ¢ncia dos Atributos - Random Forest')
plt.xlabel('Atributo')
plt.ylabel('ImportÃ¢ncia')
plt.show()
```

A coluna recÃ©m criada "Log Fare" apresenta maior correlaÃ§Ã£o com a variÃ¡vel alvo "Survived" e foi dada como a segunda mais importante para a seleÃ§Ã£o de atributos por Random Forest.
Desta forma, a coluna "Fare Interv" serÃ¡ excluÃ­da.

```python
X.drop(["Fare_Interv", "Fare"], axis=1, inplace=True)
X_teste.drop(["Fare_Interv", "Fare"], axis=1, inplace=True)
```

# SeleÃ§Ã£o e treinamento de modelos para classificaÃ§Ã£o

- RegressÃ£o logÃ­stica
- SVM
- Random Forests

```python
# Separando em treino e validaÃ§Ã£o
from sklearn.model_selection import train_test_split

X_treino, X_validacao, y_treino, y_validacao = train_test_split(X, y, test_size = 0.33, random_state=42)
```

```python
# RegressÃ£o logÃ­stica
from sklearn.linear_model import LogisticRegression

modelo1 = LogisticRegression(random_state=42, max_iter=250)
modelo1.fit(X_treino, y_treino)
pred1 = modelo1.predict(X_validacao)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
modelo2 = RandomForestClassifier(criterion="entropy", random_state=42)
modelo2.fit(X_treino, y_treino)
pred2 = modelo2.predict(X_validacao)
```

```python
# SVM
from sklearn.svm import SVC
modelo3 = SVC(C=15, kernel='rbf', tol= 0.0001, max_iter=2000, random_state=42)
modelo3.fit(X_treino, y_treino)
pred3 = modelo3.predict(X_validacao)
```

# AvaliaÃ§Ã£o dos modelos

```python
# Verificando a acurÃ¡cia dos modelos

from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix

# AcurÃ¡cia da RegressÃ£o LogÃ­stica
acc_mod1 = accuracy_score(y_validacao, pred1)

# AcurÃ¡cia da Random Forest
acc_mod2 = accuracy_score(y_validacao, pred2)

# AcurÃ¡cia do SVM
acc_mod3 = accuracy_score(y_validacao, pred3)

acc_mod1, acc_mod2, acc_mod3
```

```python
print(classification_report(y_validacao, pred1))
```

```python
print(classification_report(y_validacao, pred2))
```

```python
print(classification_report(y_validacao, pred3))
```
